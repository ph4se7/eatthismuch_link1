```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
```

```python
url = 'https://www.eatthismuch.com/recipe/nutrition/protein-southwest-scramble,254564/'
```

```python
chromedriver = r"C:\Users\User1\Downloads\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(executable_path = chromedriver)
driver.get(url)
driver.maximize_window()
driver.implicitly_wait(20)
```

```python
ingredients = driver.find_elements(By.XPATH, "//div[contains(@class, 'print_name')]")
```

```python
ing = []
for ingredient in ingredients:
    print(ingredient.text)
    ing.append(ingredient.text)
```

```python
sub_ingredients = driver.find_elements(By.XPATH, "//div[contains(@class,'print_desc food_description')]")
```

```python
sub_ing = []
for sub_ingredient in sub_ingredients:
    print(sub_ingredient.text)
    sub_ing.append(sub_ingredient.text)
```

```python
title = driver.find_element(By.XPATH, "//h2[contains(@class,'modal-title')]")
```

```python
print(title.text)
title_lst = []
title_lst.append(title.text)
```

```python
image = driver.find_element(By.XPATH, "//img[contains(@class,'lazyload')]").get_attribute("src")
```

```python
print(image)
image_url_lst = []
image_url_lst.append(image)
```

```python
recipe_steps = driver.find_element(By.XPATH, "//div[contains(@class,'directions_list collection-list')]")
```

```python
print(recipe_steps.text)
recipe = []
recipe.append(recipe_steps.text.replace("\n", " "))
```

```python
prep_time = driver.find_element(By.XPATH, "//div[contains(@class,'row text-small text-medium-md')]")
```

```python
print(prep_time.text)
prep_time_lst = []
prep_time_lst.append(prep_time.text.replace("\n", " "))
```

```python
ingredients_quantity = driver.find_elements(By.XPATH, "//div[contains(@class,'food_units col-12')]")
```

```python
quantity_lst = []
for quantity in ingredients_quantity:
    print(quantity.text)
    quantity_lst.append(quantity.text)    
```

```python
ingredients_quantity_gram = driver.find_elements(By.XPATH, "//div[contains(@class,'gram_printout col-12')]")
```

```python
gram_lst = []
for gram in ingredients_quantity_gram:
    print(gram.text)
    gram_lst.append(gram.text)
```

```python
stats = driver.find_element(By.XPATH, "//div[contains(@class,'stats_output')]")
```

```python
print(stats.text)
stats_lst = []
stats_lst.append(stats.text.replace("\n", " "))
```

```python
nutri_button = driver.find_element(By.XPATH, "//button[contains(@class,'expand_modal_nutrition btn btn-sm btn-block btn-outline-secondary top_margin_small')]")
webdriver.ActionChains(driver).click(nutri_button).perform()
detailed_nutrition = driver.find_element(By.XPATH, "//div[contains(@class,'add_micronutrients_here')]")
```

```python
print(detailed_nutrition.text)
nutrition = []
nutrition.append(detailed_nutrition.text.replace("\n", " "))
```

```python
ing_info = []
hover = driver.find_elements(By.XPATH, "//div[contains(@class,'food_box ingredient_box col-12 read-only-ingredient')]")
for i in hover:
    webdriver.ActionChains(driver).move_to_element(i).perform()
    hover_info = driver.find_element(By.XPATH, "//div[contains(@class,'tt_macro_div')]")
    ing_info.append(hover_info.text.replace("\n", " "))
    print(hover_info.text)
```

```python
a = {'Title': title_lst, 'Image URL': image_url_lst, 'Time Taken': prep_time_lst, 'Statistics': stats_lst, 'Detailed Nutrition': nutrition, 'Main Ingredients': ing, 'Sub Ingredients': sub_ing, 'Quantity': quantity_lst, 'Amount': gram_lst, 'Additional Information': ing_info, 'Recipe Steps': recipe}
```

```python
df = pd.DataFrame.from_dict(a, orient='index')
df.transpose()
```

```python
df.to_csv('data1.csv')
```
